## 一个简单的例子

Matplotlib 通过 Figure 对象绘制你的数据，每一个 Figure 对象包含一个或多个 Axes，一个区域上的点可以用多种坐标系指定。创建一个带轴线图表的最简单的方式是使用 pyplot.subplots。我们可以使用 Axes.plot 去在轴线上绘制数据：

```python
fig, ax = plt.subplots() # 创建一个包含单个轴的图像
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]) # 在轴线上绘制数据
```

## 绘图函数的输入类型

绘图函数可以接受 numpy.array 或者 numpy.ma.masked_array 作为输入，或者可以传递给 numpy.asarray 的对象。类数组的类比如 pandas 数据对象和 numpy.matrix 并不支持。通用的惯例是把他们转换成 numpy.array 去绘制。如下所示：

```python
b = np.matrix([[1, 2], [3, 4]])
b_asarray = np.asarray(b)
```

许多方法也会加载像字典一样的可寻址对象，比如 numpy.recarray 或者 pandas.DataFrame。Matplotlib 允许你提供 data 关键字作为参数去生成图表。

```python
np.random.seed(19680801)
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)
       }
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data='data')
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

## 编程风格
### 对象继承和 pyplot 接口

就像在上面提到的，有两种方式去使用 matplotlib:

- 显式创建图像和轴，并且调用他们的方法（面向对象风格）
- 依赖 pyplot 去自动创建和管理图像和轴，并且使用 pyplot 函数去绘制

面向对象风格：

```python
x = np.linspace(0, 2, 100) # 生成数据

# 注意： 在面向对象风格下，我们使用 `.pyplot.figure` 去绘制
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear') # 绘制数据
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()
```

或者 pyplot 风格

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```
